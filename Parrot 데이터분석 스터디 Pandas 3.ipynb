## 1. 데이터 합치기

### 1) concat
df을 수직 또는 수평으로 연결해줌

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7']})
```

```python
# 수직으로 연결 (행 방향으로)
result_vertical = pd.concat([df1, df2]) #디폴트는 수직으로 연결
result_vertical
```

```python
# 수평으로 연결 (열 방향으로)
result_horizontal = pd.concat([df1, df2], axis=1)
result_horizontal
```

### 2) merge
겹치는 부분이 있을 때 두 데이터를 합쳐준다

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value1': [1, 2, 3, 4]})

df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value2': [5, 6, 7, 8]})
```

```python
# 'key' 열을 기준으로 두 데이터프레임 병합
merged_df = pd.merge(df1, df2, on='key') #on에 겹치는 부분을 써줘야 함
merged_df                                #겹치는 데이터만 합쳐줌
```

## 3. 열 재설정

### 1) reindex
* labels: 재색인하려는 새로운 인덱스나 열 이름을 지정합니다.
* index: 행 인덱스를 재설정하거나 변경합니다.
* columns: 열 이름을 재설정하거나 변경합니다.
* axis: 행 인덱스 또는 열 인덱스를 변경할 축을 지정합니다. 0은 행 인덱스, 1은 열 인덱스를 나타냅니다.
* method: 재색인 시 누락된 값을 처리하는 방법을 지정합니다.
* fill_value: 누락된 값을 채우는데 사용할 기본값을 지정합니다.
* copy: 기본적으로 True로 설정되어 있으며, 새로운 객체를 생성할지 여부를 지정합니다.

```python
mpg.head()
```

```python
#열 순서 바꾸기
a=mpg[['manufacturer', 'model']]
a.reindex(columns=['model', 'manufacturer'])
```

## 4. df 정렬 및 인덱스 초기화

### 1) sort_values
