### mpg 데이터 설명
* manufacturer: 자동차의 제조사 이름
* model: 자동차 모델의 이름
* displ: 자동차의 엔진 배기량. 보통 리터(L) 단위
* year: 자동차의 제조 년도
* cyl: 자동차 엔진의 실린더 수
* trans: 자동차의 변속기 유형. 자동(Automatic), 수동(Manual)
* drv: 자동차의 구동 방식. 'F'는 전륜구동(Front), 'R'은 후륜구동(Rear), '4'는 4륜구동(4x4) 등
* cty: 도시에서의 연비.
* hwy: 고속도로에서의 연비
* fl: 연료 유형. 'r'은 정규 휘발유(Regular), 'p'은 고급 휘발유(Premium), 'd'은 디젤(Diesel) 등
* category: 자동차의 카테고리

## 2. 데이터 타입 확인 및 변환

### 1) dtypes
* df의 열별 데이터 타입을 알려줌. (메소드가 아니라 함수임!!)
* 우리 눈으로 보이는 것과 실제 데이터 타입이 다른 경우가 많음! -> 뭔가 원하는대로 잘 안되면 데이터 타입 확인해보기
* (ex. integer로 생각하고 .mean()을 했는데 오류가 났네 -> 데이터 타입 확인해보면 object일 수도 있음)
