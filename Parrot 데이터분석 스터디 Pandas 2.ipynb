## 1. value의 고유값과 그룹화

### 1) unique
각 열의 고유값을 보여줌

```python
mpg['manufacturer'].unique()
```

### 2) nunique
각 열의 고유값의 개수를 세줌

```python
mpg['manufacturer'].nunique()
```

### 3) value_counts
카테고리형 변수의 각 범주별로 개수를 세어줌

```python
mpg['manufacturer'].value_counts()
```

### 4) groupby
하나 이상의 열을 기준으로 DF를 그룹화하고 연산 (연산은 데이터분석 스터디1 파일 참고)

```python
mpg.groupby('manufacturer').mean()
```

```python
mpg.groupby(['manufacturer', 'year']).mean() #여러 열을 기준으로도 groupby 가능
```

## 2. 부울린 인덱싱

```python
#특정 조건을 만족하는 행 필터링
titanic[titanic['Age']>30]
```

```python
#여러 조건을 만족하는 행 필터링 (&:and, |:or)
titanic[(titanic['Age']>30) & (titanic['Embarked']=='S')]
```

## 3. loc, iloc을 이용한 인덱싱과 슬라이싱
* loc: 레이블 기반 인덱싱
* iloc: 정수 기반 인덱싱

### 1-1) loc을 이용한 특정 행 인덱싱 및 슬라이싱

```python
mpg.loc[0]
```

```python
mpg.loc[[0, 1, 2]]
```

```python
mpg.loc[0:2]
```

### 1-2) loc을 이용한 특정 열 인덱싱

```python
mpg.loc[:, 'cty']
```

```python
mpg.loc[:, ('cty', 'hwy')]
```

### 1-3) loc을 이용한 특정 값 접근

```python
mpg.loc[0:2, ('cty', 'hwy')]
```

### 2-2) loc을 이용한 특정 열 인덱싱

```python
mpg.iloc[:, 1:4]
```

## 4. 특정 행/열 제거하기

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 22, 35, 28]}
df = pd.DataFrame(data)
df
```

### 1) drop
* 행 제거: aixs=0
* 열 제거: aixs=1

df이 2차원이기 때문에 axis=0, 1로 처리, 만약 데이터가 3차원 이상인 경우 axis=2, 3, ...으로 확장 가능

```python
#특정 행 제거: axis=0
df.drop(2, axis=0) #인덱스가 2인 행 제거
```

```python
#특정 열 제거: aixs=1
df.drop('Age', axis=1)
```

### 2) dropna
결측치가 있는 행을 제거

```python
#결측치 확인
titanic.isna().sum()
```

```python
titanic=titanic.dropna()
```
